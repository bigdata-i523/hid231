# Seaborn Visualizations using Iris data
Here we experiment some visualizationson iris data using the seaborn package from python

```python
import seaborn as sns
```

```python
import pandas as pd

# read and import data from csv file
iris = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
# checking data size
iris.shape
```

```python
# check data
iris.head()
```

```python
iris["species"].value_counts()
```

```python
# display the visualization below the code itself
%matplotlib inline
```

# Visualization 1 : Joint plot for sepal_length and sepal_width:

```python
# This is a joint plot for two variables sepal_length and sepal_width
sns.jointplot(x ="sepal_length", y = "sepal_width", data =iris, size= 5)
```

# Visualization 2: Joint plot for petal_length and petal_width:

```python
# This is a joint plot for two variables petal_length and petal_width
sns.jointplot(x ="petal_length", y = "petal_width", data =iris, size= 5)
```

# Visualization 3: Facet grid Plots

```python
import matplotlib.pyplot as plt
sns.FacetGrid(iris, hue = "species", size =5)\
    .map(plt.scatter, "sepal_length", "sepal_width")\
    .add_legend()
```

# Visualization 4: Boxplot for petal_length

```python
ax = sns.boxplot(x = "species", y = "petal_length", data = iris)
```

# Visualization 5: Boxplot for petal_width

```python
ax = sns.boxplot(x = "species", y = "petal_width", data = iris)
```

# Visualization 6: Boxplot for sepal_length

```python
ax = sns.boxplot(x = "species", y = "sepal_length", data = iris)
```

# Visualization 7: Boxplot for sepal_width

```python
ax = sns.boxplot(x = "species", y = "sepal_width", data = iris)
```

# Visualization 8: Pair plots

```python
sns.set(style="ticks", color_codes=True)
g = sns.pairplot(iris, kind="scatter")
```
